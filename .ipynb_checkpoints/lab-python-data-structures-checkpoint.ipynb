## Exercise: Managing Customer Orders

As part of a business venture, you are starting an online store that sells various products. To ensure smooth operations, you need to develop a program that manages customer orders and inventory.

Follow the steps below to complete the exercise:

1. Define a list called `products` that contains the following items: "t-shirt", "mug", "hat", "book", "keychain".

2. Create an empty dictionary called `inventory`.

3. Ask the user to input the quantity of each product available in the inventory. Use the product names from the `products` list as keys in the `inventory` dictionary and assign the respective quantities as values.

4. Create an empty set called `customer_orders`.

5. Ask the user to input the name of three products that a customer wants to order (from those in the products list, meaning three products out of "t-shirt", "mug", "hat", "book" or "keychain". Add each product name to the `customer_orders` set.

6. Print the products in the `customer_orders` set.

7. Calculate the following order statistics:
   - Total Products Ordered: The total number of products in the `customer_orders` set.
   - Percentage of Products Ordered: The percentage of products ordered compared to the total available products.
   
   Store these statistics in a tuple called `order_status`.

8. Print the order statistics using the following format:
   ```
   Order Statistics:
   Total Products Ordered: <total_products_ordered>
   Percentage of Products Ordered: <percentage_ordered>% 
   ```

9. Update the inventory by subtracting 1 from the quantity of each product. Modify the `inventory` dictionary accordingly.

10. Print the updated inventory, displaying the quantity of each product on separate lines.

Solve the exercise by implementing the steps using the Python concepts of lists, dictionaries, sets, and basic input/output operations. 

````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
inventory = {}

user_input1 = input("How many t-shirts are there? ")
user_input = True
while True:
    try:
        user_input_n = int(user_input1)
        break
    except ValueError:
        print("Please enter a valid number")
        user_input1 = (input("How many t-shirts are there? "))
inventory["t-shirt"] = int(user_input1)

user_input2 = input("How many mugs are there? ")
while True:
    try:
        user_input_n = int(user_input2)
        break
    except ValueError:
        print("Please enter a valid number")
        user_input2 = (input("How many mugs are there? "))
inventory["mug"] = int(user_input2)

user_input3 = input("How many hats are there? ")
while True:
    try:
        user_input_n = int(user_input3)
        break
    except ValueError:
        print("Please enter a valid number")
        user_input3 = (input("How many hats are there? "))
inventory["hat"] = int(user_input3)

user_input4 = input("How many books are there? ")
while True:
    try:
        user_input_n = int(user_input4)
        break
    except ValueError:
        print("Please enter a valid number")
        user_input4 = (input("How many books are there? "))
inventory["book"] = int(user_input4)

user_input5 = input("How many keychains are there? ")
while True:
    try:
        user_input_n = int(user_input5)
        break
    except ValueError:
        print("Please enter a valid number")
        user_input5 = (input("How many keychains are there? "))
inventory["keychain"] = int(user_input5)

customer_orders = set()
print("This are our products: ", products)

while len(customer_orders) <= 3:  # we are going to add elements in a set. Elements are UNIQUE so they are counted only once even if the customer enters twice the same product!
    customer_input = input("Please enter a product you would like to order. But remember, you can order max 3 types of product: ")
  
    if customer_input in inventory:
        if inventory[customer_input] > 0:
            if customer_input not in customer_orders:
                customer_orders.add(customer_input)
                print("This product was added to you order!")
                inventory[customer_input]-= 1
            
        elif customer_input in customer_orders:
            print("You have already order this product. Please order a new one: ")
        else:
            print("Sorry, this product is out of stock.")       

    else:
        print("Sorry, we don't have this product.")
        
    if len(customer_orders) == 3:
        break

print(f"This is the list of products you ordered: {customer_orders}")

total_order = len(customer_orders)
total_inventory = sum(inventory.values())
percentage_order = (total_order/total_inventory)*100  # % of ordered products in the inventory
order_status = (total_order, percentage_order) # tuple storing the statistics of ordered products

print(f"""Order Statistics: 
      Total Products Ordered:{total_order}
      Percentage of Products Ordered: {percentage_order}%""")
      
print("The inventory is now: ")
for key, value in inventory.items():
    print(f"{key}: {value}")
````
